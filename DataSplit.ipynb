```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split

# Set the paths
dataset_root = r'C:/Users/user/Downloads/Plantdata/Dataset'
train_dir = r'C:/Users/user/Desktop/Organized_data/train_set'
val_dir = r'C:/Users/user/Desktop/Organized_data/validation_set'
test_dir = r'C:/Users/user/Desktop/Organized_data/test_set'

# Create directories
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Get class folders
class_folders = [class_folder for class_folder in os.listdir(dataset_root) if os.path.isdir(os.path.join(dataset_root, class_folder))]

# Iterate through each class folder
for class_folder in class_folders:
    class_path = os.path.join(dataset_root, class_folder).replace('\\', '/')

    # Split the images into train, validation, and test sets
    images = [img for img in os.listdir(class_path) if img.endswith(".jpg")]
    train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
    train_images, val_images = train_test_split(train_images, test_size=0.1, random_state=42)

    # Create directories for the current class in train, validation, and test sets
    os.makedirs(os.path.join(train_dir, class_folder), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_folder), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_folder), exist_ok=True)

    # Copy images to their respective directories Training set: 70%, Validation set: 10%, Test set: 20%
    for img in train_images:
        src_path = os.path.join(class_path, img)
        dest_path = os.path.join(train_dir, class_folder, img)
        shutil.copy(src_path, dest_path)

    for img in val_images:
        src_path = os.path.join(class_path, img)
        dest_path = os.path.join(val_dir, class_folder, img)
        shutil.copy(src_path, dest_path)

    for img in test_images:
        src_path = os.path.join(class_path, img)
        dest_path = os.path.join(test_dir, class_folder, img)
        shutil.copy(src_path, dest_path)
```

```python
import os

# Set the paths
train_dir = r'C:/Users/user/Desktop/Organized_data/train_set'
val_dir = r'C:/Users/user/Desktop/Organized_data/validation_set'
test_dir = r'C:/Users/user/Desktop/Organized_data/test_set'

# Function to count images in each class
def count_images_per_class(directory):
    class_counts = {}
    for class_folder in os.listdir(directory):
        class_path = os.path.join(directory, class_folder)
        if os.path.isdir(class_path):
            num_images = len([img for img in os.listdir(class_path) if img.endswith(".jpg")])
            class_counts[class_folder] = num_images
    return class_counts

# Count images in each class for training set
train_class_counts = count_images_per_class(train_dir)
print("Training Set:")
for class_name, num_images in train_class_counts.items():
    print(f"{class_name}: {num_images} images")

# Count images in each class for validation set
val_class_counts = count_images_per_class(val_dir)
print("\nValidation Set:")
for class_name, num_images in val_class_counts.items():
    print(f"{class_name}: {num_images} images")

# Count images in each class for test set
test_class_counts = count_images_per_class(test_dir)
print("\nTest Set:")
for class_name, num_images in test_class_counts.items():
    print(f"{class_name}: {num_images} images")
```
